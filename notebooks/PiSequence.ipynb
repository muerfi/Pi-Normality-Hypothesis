# Finite-prefix exploration of π digits

This notebook is a lightweight companion to the `pi_lab` command-line tools. It is intended for reproducible exploration of finite prefixes of the decimal expansion of π. It does not prove normality of π.


## Interpretation rule

Every result in this notebook is about the loaded finite digit file. Frequency summaries are finite descriptive statistics. Target-string searches are exact searches over the loaded prefix only. Occurrence does not carry interpretive content, and non-occurrence in the loaded prefix does not imply non-occurrence in π.


```python
from pathlib import Path
from pi_lab.digits import load_pi_digits
from pi_lab.search import search_sequence
from pi_lab.statistics import block_frequencies

digits = load_pi_digits(Path('../data/pi_decimals.txt') if Path('../data/pi_decimals.txt').exists() else None)
len(digits)
```

## Single-digit counts

Single-digit counts are a useful first check, but normality requires the limiting frequencies of every finite block length.


```python
block_frequencies(digits, block_size=1)
```

## Target-string search

The next cell searches for the first occurrence of `314159` in the loaded prefix. Because the project stores the leading `3` and omits the decimal point, this target is expected at zero-based position 0 when the file begins with π.


```python
search_sequence('314159', digits)
```
